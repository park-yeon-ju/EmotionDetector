# 4. Real Time Detections

```python
# webcam에 접근 및 real time detections을 만들 수 있도록 해주는 코드
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    # 이 밑으로는 이쁘게 프로그램을 종료하는 코드
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# 5. Train from scratch

```python
import uuid  # unique identifier / name the images
import os
import time
```

```python
IMAGES_PATH = os.path.join('data','images')

# 촬영할 이미지 카테고리
labels = ['happy', 'surprise','angry','sad'] 

# label 별 촬영할 이미지 개수
number_imgs = 10 
```

```python
cap = cv2.VideoCapture(0)

# Loop through label range
for label in labels:
    print('collecting images for {}'.format(label))
    print('collecting images for {}'.format(label))
    print('collecting images for {}'.format(label))
    time.sleep(10)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label +'.'+str(img_num + 10)+'.jpg')
        
        # Writes out image to file
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(5)
        
        # 이 밑으로는 이쁘게 끝내는 코드
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
