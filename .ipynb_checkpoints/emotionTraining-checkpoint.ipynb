# 4. Real Time Detections

```python
# webcamÏóê Ï†ëÍ∑º Î∞è real time detectionsÏùÑ ÎßåÎì§ Ïàò ÏûàÎèÑÎ°ù Ìï¥Ï£ºÎäî ÏΩîÎìú
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    # Ïù¥ Î∞ëÏúºÎ°úÎäî Ïù¥ÏÅòÍ≤å ÌîÑÎ°úÍ∑∏Îû®ÏùÑ Ï¢ÖÎ£åÌïòÎäî ÏΩîÎìú
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# 5. Train from scratch

```python
import uuid  # unique identifier / name the images
import os
import time
```

```python
IMAGES_PATH = os.path.join('data','images')

# Ï¥¨ÏòÅÌï† Ïù¥ÎØ∏ÏßÄ Ïπ¥ÌÖåÍ≥†Î¶¨
labels = ['happy', 'surprise','angry','sad'] 

# label Î≥Ñ Ï¥¨ÏòÅÌï† Ïù¥ÎØ∏ÏßÄ Í∞úÏàò
number_imgs = 10 
```

```python
cap = cv2.VideoCapture(0)

# Loop through label range
for label in labels:
    print('collecting images for {}'.format(label))
    print('collecting images for {}'.format(label))
    print('collecting images for {}'.format(label))
    time.sleep(10)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label +'.'+str(img_num + 10)+'.jpg')
        
        # Writes out image to file
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(5)
        
        # Ïù¥ Î∞ëÏúºÎ°úÎäî Ïù¥ÏÅòÍ≤å ÎÅùÎÇ¥Îäî ÏΩîÎìú
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
